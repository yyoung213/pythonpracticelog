찾아낸 규칙
1. 주어진 수는 column의 개수이지만 출력이 행단위로 된다는 파이썬 언어의 특성상 column의 개수와 row의 개수와의 관계를 찾음
 - column의 개수 = 2 * row의 개수 - 1
2. row를 기준으로 출력되는 '*'의 개수가 하나씩 증가하다가 주어진 column번째 row 부터는 한개씩 감소함


#규칙이 바뀌는 지점에 주목한 코드

for문과 if문을 이용한 코드
- if-else문을 사용하여 규칙이 바뀌는 지점인 ncol번째 row에서 cnt의 증감이 바뀌도록 하였습니다.

````python
ncol = int(input('number of columns: '))
nrow = 2*ncol - 1  #행의 개수
cnt = 0 #출력되는 *의 개수 

for i in range(1, nrow+1):  #반복문의 시행 횟수 = 행의 개수 , 따라서 i번째 행에 대한 식
    if i <= ncol:           #시행횟수가 입력된 열의 개수와 같을때까지
        cnt += 1            #cnt를 하나씩 증가해가며 *을 출력
        print('{}'.format('*'*cnt))
    else:                   #출력된 *의 개수 = 열의 개수가되면
        cnt -= 1            #cnt가 하나씩 감소해가며 *을 출력
        print('{}'.format('*'*cnt))
        
````

위 코드의 장점은 직관적으로 이해하기 쉽다는 것이지만, 불필요한 변수가 존재하고 for문에서 매번 조건문을 사용하다보니 속도가 느려질수 있다는 단점이 있습니다.

코드를 작성하고 관찰하여보니 ncol번째 행 이후에는 행과 별의 개수의 합이 2*ncol로 일정하다는 것을 알수 있었고, 이를 코드로 표현하였습니다.
  
- 2 * 열의 개수 = 행의 순번 + 출력되는 별의 개수

````python
ncol = int(input('number of columns: '))
nrow = 2*ncol   #행의 개수 + 1           

for i in range(1, nrow):
    if i <= ncol:
        print('{}'.format('*'*i))
    else:
        print('{}'.format('*'*(nrow-i)))
````

````python
ncol = int(input('number of columns: '))
nrow = 2*ncol  #행의 개수 + 1
mark = '*'  #별표 대신 다른 기호로 변경이 용이

for i in range(1, nrow):
    if i <= ncol: 
        print('{}'.format(mark*i))
    else:
        print('{}'.format(mark*(nrow-i)))
````

if 문 사용에 대한 개선

for문 연속하여 사용하여 출력하기
- if문을 사용할 필요가 없기 때문에 속도에 있어서 강점을 보일것으로 생각됨

````python
ncol=int(input("Number of Columns: "))
mark = '*'

for i in range(1,ncol):
    print('{}'.format(mark*i))
for i in range(ncol,0,-1):
    print('{}'.format(mark*i))
````

#대칭성에 초점을 맞춰 작성한 코드
- 행의 개수는 항상 ncol번째 행에 대해서 대칭성을 지님. 따라서 수직선상에서 평행이동할수 있음
- column의 개수는 별의 최대 개수와 같음

절대값내장함수 abs()를 이용한 코드



````python
ncol=int(input('number of columns: '))
mark = '*'

for i in range(1-ncol,ncol):    
    print('{}'.format(mark*(ncol-abs(i))))
````

#규칙을 리스트에 저장하여 출력하기
- 장점 : 규칙이 바뀌었을때 코드를 수정하기 용이함

```
mark = 기호
list = 각 행에서 출력되어야할 기호의 개수

for i in list:
    print(mark*list)
```



````python
ncol=int(input("Number of Columns: "))
mark = '*'
row_list = list(range(1,ncol))+list(range(ncol,0,-1))
for i in row_list:
    print('{}'.format(mark*i))
````

````python
ncol=int(input("Number of Columns: "))
mark = '*'
row_list = list(map(abs,range(1-ncol,ncol))) #abs()와 map()을 이용하여 규칙 list 구성

for i in row_list:
    print('{}'.format(mark*(ncol-i)))
````
