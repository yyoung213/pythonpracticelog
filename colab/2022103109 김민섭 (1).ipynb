```python
score = [[55, 30, 85, 70, 90], [74, 65, 86, 79, 100],  [80, 90, 100, 71, 62]]
name = ['김대한','김민국','김선문','김한국','김무궁']

for i in range(len(name)):
  sum_score = 0
  for j in range(len(score)):
    sum_score += score[j][i]
  print(name[i], "-- > [총점] : ", sum_score, "[평균] : ", round(sum_score/len(score), 2))
```
