```python
sentence="I would like to attend X in the next year.               "
conf="KDD"

new_sentence=sentence.strip()

idx=new_sentence.find('X')
new_sentence=new_sentence[:idx] +conf+new_sentence[idx+1:]

print("My sentence:",new_sentence)
print("Old sentence Length:",len(sentence))
print("New sentence Length:",len(new_sentence))
```

```python
#문자열 나누기split 함수
line='hello| you| world| and universe!\n'
words=line.split()
print(words)
words=line.split('|')
print(words)
words=line.split('|',1)
print(words)
words=line.split('|',2)
print(words)
words=line.split('|')[2]
print(words)
```

```python
#문자열 결합join 함수
colors=['red','blue','green','yellow']
result=''.join(colors)
print(result)
result=' '.join(colors)
print(result)
result=','.join(colors)
print(result)
result='-'.join(colors)
print(result)
```

```python
sentence="I would like to attend X in the next year.               "
conf="KDD"
n_sentence=sentence.split()
n_sentence[n_sentence.index('X')]=conf
n_sentence=' '.join(n_sentence)


print("My sentence:",n_sentence)
print("Old sentence Length:",len(sentence))
print("New sentence Length:",len(n_sentence))
```
