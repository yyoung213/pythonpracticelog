```python
birth=int(input("Enter your birth: "))
age=(2022-birth+1)
if age<=26 and age>=20:
    print('대학생')
elif age<20 and age>=17:
    print('고등학생')
elif age<17 and age>=14:
    print('중학생')
elif age<14 and age>=8:
    print('초등학생')
else:
    print('학생이 아닙니다.')
```

for 변수 in 리스트:

리스트의 변수가 n개일때 아래실행문을 n번실행

```python
for looper in [1,2,3,4,5]:
    print("hello")
```
