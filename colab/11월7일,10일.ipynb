```python
value=int(input('Number of Rows: '))
star='*'
for i in range(value):
    star="*"*(i*2+1)
    print(f'{star:^20s}')
```

```python
ncol=int(input("Number of Columns: "))
nrow = ncol-1 
format_str = '{:<'+str(nrow)+'}'
for i in range(1,nrow-1):
    print(format_str.format("*"*i))
for i in range(nrow-1,0,-1):
    print(format_str.format("*"*i))
```

```python
num = int(input('number of columns: '))
nrow = 2*num - 1
cnt = 1
cnt_2 = num
while cnt <= nrow:
    if cnt <= num:
        print('*'*cnt)
        cnt += 1
    else:
        cnt_2 -= 1
        print('*'*cnt_2)
        cnt += 1
```
