```python
sentence = list(input('문장을 자유롭게 입력하세요.>> \n').lower())
special = ['\"',"\'",'~','`','@','#','!','$','%','^','&','*','(',')','_','-','+','=','{','[','}',']','|','\\','<',',','>','.','?','/',':',';']
for i in range(len(sentence)):
    if sentence[i] in special:
        sentence[i] = ''
sentence_new = ''.join(sentence).split()
word = sorted(list(set(''.join(sentence).split())))
print('-'*40)
print(f'{"단어":^18s}\t{"빈도":^10s}')
print('-'*40)
for i in word:
    cnt = 0
    for  j in sentence_new:
        if i==j:
            cnt += 1
    print(f'{i:^20s}\t{cnt:^12d}')
```

```python
#스택과 큐


a=[1,2,3,4,5]
a.append(10)
a
a.append(20)
a
a.pop()
a.pop()

building=[]
vehicle=[]
while True:
    name=input('이름을 입력해주세요(마지막이면 q/Q 입력):')
    building.append(name)
    if name in ['Q'or 'q']:
        break

print('Stack'.center(50,'-'))
print('Building:{}\nVehicle:{}'.format(building,vehicle))

while building:
    name=building.pop()
    vehicle.append(name)

print('Pop'.center(50,'-'))
print('Building:{}\nVehicle:{}'.format(building,vehicle))
```

```python
word=list(input('input a word:'))
stack=[]
queue=[]
print('Initial:{}'.format(word))
while word:
    s_word=word.pop()
    stack.append(s_word)
    
print('==> By stack:{}'.format(stack))
while stack:
    queue.insert(0,stack.pop(0))
    
print('==> By Queue:{}'.format(queue))
```
