세트

값을 순서없이 저장하면서 중복을 불허하는 자료구조

set()

```python
#s1={3,1,2,1,2,4}
#s2=set([1,2,3,1,2,3])#리스트의 셋변환
#s3=set((1,2,3)*2)
#print(s1)
#print(s2)
#print(s3)

#내장함수
#add() remove()해당 원소가 없으면 에러 discard()해당 원소가 없어도 ㄱㅊ update() clear()

s=set([1,2,3,1,2,3])
print(s)
s.add(1)
print(s)
s.remove(1)
print(s)
s.update([1,4,5,6,7])
print(s)
s.discard(3)
print(s)
s.clear()
print(s)

#합집합
s1=set([1,2,3,4,5])
s2=set([3,4,5,6,7])

print(s1.union(s2))
print(s1 | s2)
#교집합
s1=set([1,2,3,4,5])
s2=set([3,4,5,6,7])

print(s1.intersection(s2))
print(s1 & s2)
#차집합

s1=set([1,2,3,4,5])
s2=set([3,4,5,6,7])

print(s1.difference(s2))
print(s1-s2)

print(s2.difference(s1))
print(s2-s1)

```
