```python
from nltk.stem import PorterStemmer#어간 추출
ps= PorterStemmer()
words=['likes','liked','likely']
for w in words:
    print(f'[word]:{w:6s}, [stem]:{ps.stem(w)}')
```

```python
from nltk.stem import WordNetLemmatizer#표제어 추출
nltk.download('wordnet')
nltk.download('omw-1.4')

w1=WordNetLemmatizer()
words={'v':['is','was','are','were','being'],#verbs:'v'
       'a':['better','smaller','smallest','bigger','largest']}#adverb:'a'

for pos in words.keys():
    for word in words[pos]:
        print(f'[word]:{word:8s},[lemmatization]:{w1.lemmatize(word,pos=pos)}')
```
