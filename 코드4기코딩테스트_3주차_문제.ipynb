**문제2**
=============

1부터 6까지 숫자가 적힌 주사위가 세 개 있습니다. 세 주사위를 굴렸을 때 나온 숫자를 각각 a, b, c라고 했을 때 얻는 점수는 다음과 같습니다.

- 세 숫자가 모두 다르다면 a + b + c 점을 얻습니다.
- 세 숫자 중 어느 두 숫자는 같고 나머지 다른 숫자는 다르다면 (a + b + c) × (a2 + b2 + c2 )점을 얻습니다.
- 세 숫자가 모두 같다면 (a + b + c) × (a2 + b2 + c2 ) × (a3 + b3 + c3 )점을 얻습니다.
세 정수 a, b, c가 매개변수로 주어질 때, 얻는 점수를 return 하는 함수를 작성해 주세요.

- 함수 이름: Dice_score
- 입력 파라메터: 1~6사이의 정수 3개

```python
#답안 작성#
def Dice_score(a,b,c):
    a,b,c=map(int,input().split())
    if a == b == c:
        return (a+b+c)*(a**2+b**2+c**2)*(a**3+b**3+c**3)
    if a !=b != c:
        return a+b+c
    else:
        return (a+b+c)*(a**2+b**2+c**2)
result=Dice_score('','','')
print(result)
```

- 함수 이름:    Cal_divisor
- 입력 파라메터:    정수 1개

```python
#답안 작성#
def Cal_divisor(n):
    list=[]
    n=int(n)
    if n>=0:
        for args in range(n):
            if n%(int(args)+1)==0:
                list.append(args+1)
    
        return(list)
    else:
        n=n*(-1)
        for args in range(n):
            if n%(int(args)+1)==0:
                p_list=list.append(args+1)
                m_list=[]

                for i in p_list:
                    m_list.append(int(i)*(-1))
                f_list=m_list.reverse()+p_list
        return(f_list)

Cal_divisor(-12)
```
