```python
def sung():
    while True:
        score=input('학생의 점수를 정수로 입력하시오(0이상 100이하):')
        if score.isdigit()==False:print('[경고메세지]: 잘못 입력하셨습니다. 다시 입력하십시오.');break
        if int(score)>=90 and int(score)<=100:
            grade='A'
            break
        elif int(score)>=80:
            grade='B'
            break
        elif int(score)>=70:
            grade='C'
            break
        elif int(score)>=60:
            grade='D'
            break
        elif int(score)>=0 and int(score)<60:
            grade='F'
            break
        else:
            print('[경고메세지]: 잘못 입력하셨습니다. 다시 입력하십시오.')
    while True:
        absent=input('학생의 결석일 수를 정수로 입력하시오.(0이상 수업일수 이하):')
        if int(absent)>=30*0.2 and int(absent)<30:
            grade='F'
            print('==> 학생의 성적등급은 F입니다.')
            break
        elif int(absent)<30*0.2:
            print('==> 학생의 성적등급은',grade,'입니다.')
            break
        else:
            print('[경고메세지]: 잘못 입력하셨습니다. 다시 입력하십시오.')
    return


while True:
    sung()
    step=input('"계속 입력하시겠습니까?(YES:y혹은 Y,NO: n혹은 N)"')
    if step in ['n','N']:
        break
    while True:
        if step in['y','Y']:
            break
        elif step!='n'and step!='N' and step!='y' and step!='Y':
            print('[경고메세지]: 잘못 입력하셨습니다. 다시 입력하십시오.')
            break    
```

```python
import random as rd
a=1
b=1
score=0
while True:
    step=input('구구단 몇단을 계산할까요(1에서 9입력, 게임종료시 0입력)?')
    num=rd.choice(range(1,9))
    if step in ['0']: print('구구단 게임을 종료합니다. 아래 결과를 확인하세요');break
    while True:
        if step.isdigit()==False: print('잘못 입력했습니다. 1부터 9사이 정수를 입력하세요')
        
        print(step,'X',num,'>>')
        answer=input()
        if int(answer)!=num*int(step):
            
            score-=int(step)
            print('오답!!: -',step,'점 ==> [누적점수]: ',score )
            break
        else:
            score+=int(step)
            print('정답!!: +',step,'점 ==> [누적점수]: ',score )
        
            break
print('[총점]: ',score)
```
