```python
def is_vowel(letter):
    vowels='aieou'
    return letter.lower() in vowels

sentence=input('문장을 입력하세요: ')
print(set([i for i in sentence.lower() if is_vowel(i)]))#피뎊보고 다시하기 
```

```python
case_1=['A','B','C']
case_2=['D','E','A']
result=[]
for j in case_2:
    ob=[]
    for i in case_1:
        ob.append(i+j)
    result.append(ob)

print(ob)
print(result)
```
