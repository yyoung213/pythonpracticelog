```python
import random as rd
a=1
b=1
while True:
    step=input('구구단 몇단을 도전하시겠습니까?')
    num=rd.choice(range(1,10))
    if step in ['q','Q']: print('--- 구구단을 종료합니다. ---');break
    while True:
        print(step,'X',num,'>>')
        
        
        answer=input()
        if int(answer)!=num*int(step):
            a += 1
            print('정답이 아닙니다. 다시 도전하세요')
        else:
            print('정답입니다!')
            b += 1
            break
win=b/(a+b)
print('[정답횟수]:',a)
print('[오답횟수]:',b)
print('[승률]:',round(win,2))
```
