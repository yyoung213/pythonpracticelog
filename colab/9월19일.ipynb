```python
cities=['서울','부산','인천','대구','대전','광주','울산','수원']
print(cities[::2])#처음부터 끝까지 간격2
print(cities[::-1])#거꾸로
print(cities[::-2])#거꾸로 2칸씩
```

```python
rainbow_part1=['빨강','주황','노랑','초록']
rainbow_part2=['파랑','남색','보라']
print(rainbow_part1 +rainbow_part2)
print('무지개(p1):',len(rainbow_part1))
print('무지개(p2):',len(rainbow_part2))

rainbow= rainbow_part1+rainbow_part2
print(rainbow)
print('무지개(p1+p2):',len(rainbow))
print('초록'in rainbow_part1)#in 연산
print('파랑'in rainbow_part1)#in연산
```

리스트는 곱셈,덧셈, in연산 가능

rainbow_part1*2   리스트 2번반복

append 새로운 값을 끝에 추가 color.append('white')
extend 새로운 리스트를 기존 리스트에 추가 color.extend['black','purple']
 insert 기존 리스트의i번째 인덱스에 새로운값을 추가color.insert(0.'orange')
 remove 리스트 내의 특정값을 삭제 color.remove('white')
  del 특정 인덱스값을 삭제

```python
planet=['금성','지구','화성']
planet.append('목성')
print(planet)
planet.extend(['토성','천왕성'])
print(planet)
planet.insert(0,'수성')
print(planet)
planet.insert(2,'경희대')
print(planet)
planet.remove('경희대')
print(planet)
planet[0]='서울'
print(planet)

del planet[0]
planet
```

#패킹과 언패킹

패킹 : 한 변수에 여러개의 데이터를 할당하는 것

언패킹 : 한 변수의 데이터를 각각의 변수로 변환하는 것

```python
t=[1,2,3]#패킹
a,b,c=t#언패킹
print(t)
print(a,b,c)
```

```python
t=[1,2,3]
a,b,c,d,e=t
print(a,b,c,d,e) #패킹이 3갠ㄴ데 언패킹할때 5개에 나눔 에러
```
