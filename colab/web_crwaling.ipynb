##https://www.musinsa.com/categories/item/003?d_cat_cd=003&brand=&list_kind=small&sort=pop_category&sub_sort=&page=1&display_cnt=90&group_sale=&exclusive_yn=&sale_goods=&timesale_yn=&ex_soldout=&kids=&color=3&price1=&price2=&shoeSizeOption=&tags=&campaign_id=&includeKeywords=&measure=

```python
start = 1
cotton_list = []


while start <= 100: #바지의 각 카테고리 페이지에 맞춰서 페이지 수 바꾸기 ex)코튼팬츠는 100페이지임으로 100으로 놔둠
    try:
        url ='https://www.musinsa.com/categories/item/003007?d_cat_cd=003007&brand=&list_kind=small&sort=pop_category&sub_sort=&page={}&display_cnt=90&group_sale=&exclusive_yn=&sale_goods=&timesale_yn=&ex_soldout=&kids=&color=&price1=&price2=&shoeSizeOption=&tags=&campaign_id=&includeKeywords=&measure='.format(start)
        #url을 카테고리에 맞춰서 바꾸기 -- 페이지 마다 page=부분을 제외한 나머지가 동일하므로 page=1,2,,,,,가 되도록 format을 이용해 url 지정정
        headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"}
        #what is my agent를 검색하여 알 수 있음 -- 웹사이트에서 사람임을 입증할 때 활용
        res = requests.get(url, headers=headers)
        res.raise_for_status()

        soup = BeautifulSoup(res.text, "lxml")
        clothesBox = soup.find_all('p', attrs={"class": "list_info"}) 
        # 사이트에서 f12누르고 화살표 클릭해보면 위의 사진에서 보이는 'list info'가 상품 이름을 나타냄을 알 수 있음음

        box = []
        for i in clothesBox:
            clothes = i.find('a') # list info가 붙은 부분에서 'a'로 시작하는 부분만 찾아서 clothes에 할당당
            box.append(clothes) 

        for clothes in box: 
            goods = clothes.get("title") # 'a'가 붙은 부분 중 'title'만 가져옴옴
            cotton_list.append(goods)
            
        start += 1 #1페이지를 시작으로 100페이지가 될 때 까지 쭉쭉
    
    except:
        print(start)
        break
print(len(cotton_list))

cotton = {}
for i in cotton_list:
    cotton[i] = 1
```

```python
def goods_dict(wear, url_user, page):
    start = 1
    goods_list = []

    while start <= page:
        try:
            url = url_user.format(start)
            headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"}
            res = requests.get(url, headers=headers)
            res.raise_for_status()

            soup = BeautifulSoup(res.text, "lxml")
            clothesBox = soup.find_all('p', attrs={"class": "list_info"}) 

            box = []
            for i in clothesBox:
                clothes = i.find('a')
                box.append(clothes) 

            for clothes in box: 
                goods = clothes.get("title") 
                goods_list.append(goods) #customizing
            
            start += 1
    
        except:
            print(start)
            break

    wear = {}
    for i in goods_list: #customizing
        wear[i] = 1 #customizing

    return wear
```
